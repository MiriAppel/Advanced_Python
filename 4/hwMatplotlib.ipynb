תרגיל 2: גרף עמודות

מטרה: להציג את כמות המכירות לפי קטגוריה

נתונים: מכירות של 4 קטגוריות מוצרים בחנות

```python
categories = ['shirt', 'shoes', 'skirt', 'dress']
values = [100, 200, 150, 300]
```

```python
plt.bar(categories, values)
plt.title('Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')
```

תרגיל 3: גרף פיזור

מטרה: לחקור את הקשר בין זמן לימוד לציונים במבחן

נתונים: שעות לימוד וציונים של תלמידים

```python
hours = [6, 2, 9, 10, 20, 1, 7, 16]
marks = [69, 88, 95, 100, 99, 78, 65, 92]
```

```python
plt.scatter(hours, marks, color='blue')
plt.title('Scatter graph')
plt.xlabel('Hours')
plt.ylabel('Marks')
```

תרגיל 4: היסטוגרמה

מטרה: לנתח את התפלגות גילאים בקבוצת נבדקים

נתונים: גילאים של אנשים בקבוצת מחקר

```python
ages = [22, 45, 30, 33, 50, 41, 28, 33, 55, 65, 29, 42, 48, 31]

plt.hist(ages, bins=[20, 30, 40, 50, 60, 70], edgecolor='black')

plt.title('Histogram')
plt.xlabel('Ages')
plt.ylabel('Number of People')
```

תרגיל 5: גרפים מרובים בחלון אחד

מטרה: להשוות בין צריכת אנרגיה לפי שנים ומקורות

נתונים: צריכת אנרגיה לפי שנים ומקורות אנרגיה שונים

```python
years = range(2010, 2021)
consumption_oil = [100, 105, 110, 120, 125, 130, 135, 140, 145, 150, 155]
consumption_gas = [80, 82, 85, 88, 90, 95, 100, 105, 110, 115, 120]
consumption_renewable = [20, 25, 30, 35, 40, 45, 50, 60, 70, 80, 90]
```

```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(years, consumption_oil, 'tab:blue')
axs[0, 0].set_title('Oil Consumption')
axs[0, 0].set_xlabel('Year')
axs[0, 0].set_ylabel('Consumption (units)')

axs[0, 1].plot(years, consumption_gas, 'tab:red')
axs[0, 1].set_title('Gas Consumption')
axs[0, 1].set_xlabel('Year')
axs[0, 1].set_ylabel('Consumption (units)')

axs[1, 0].plot(years, consumption_renewable, 'tab:green')
axs[1, 0].set_title('Renewable Consumption')
axs[1, 0].set_xlabel('Year')
axs[1, 0].set_ylabel('Consumption (units)')

axs[1, 1].plot(years, consumption_oil, 'tab:blue', label='Oil')
axs[1, 1].plot(years, consumption_gas, 'tab:red', label='Gas')
axs[1, 1].plot(years, consumption_renewable, 'tab:green', label='Renewable')
axs[1, 1].set_title('Energy Consumption')
axs[1, 1].set_xlabel('Year')
axs[1, 1].set_ylabel('Consumption (units)')
axs[1, 1].legend()

for ax in axs.flat:
    ax.label_outer()
```
