```python
import pandas as pd
```

שלב 1: טעינת נתונים

```python
df = pd.read_csv('cities.csv')
df
```

שלב 2: ניקוי נתונים

```python
df.fillna({
    'LatD': df['LatD'].mean(),
    ' "LatM"': df[' "LatM"'].mean(),
    ' "LatS"': df[' "LatS"'].mean(),
    ' "NS"': 0,
    ' "LonD"': df[' "LonD"'].mean(),
    ' "LonM"': df[' "LonM"'].mean(),
    ' "LonS"': df[' "LonS"'].mean(),
    'EW': 0,
    ' "City"': 0,
    ' "State"': 0
}, inplace=True)
```

```python
df
```

שלב 3: ניתוח וחישובים

```python
import numpy as np
```

```python
df['population'] = np.random.randint(1000, 10001, size=len(df))
df
```

```python
avg = df.groupby(' "State"')['population'].mean()
avg
```

```python
largest_city = df.loc[df['population'].idxmax()]
name = largest_city[' "City"']
name
```

```python
smallest_city = df.loc[df['population'].idxmin()]
name = smallest_city[' "City"']
name
```

שלב 4: שינויים והוספות

```python
df['Area'] = np.random.randint(10, 100, size=len(df))
df
```

```python
df.rename(columns={'population': 'Population'}, inplace=True)
```

```python
df['Population_per_km'] = df['Population'] / df['Area']
```

```python
df
```

שלב 5: סינון ומיון

```python
states = df[' "State"'].unique()
states
```
